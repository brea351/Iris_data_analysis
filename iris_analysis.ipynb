```python
# iris_analysis.py

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Task 1: Load and Explore the Dataset
try:
    # Load Iris dataset from seaborn
    iris = sns.load_dataset('iris')
    print("Dataset Loaded Successfully!\n")
except Exception as e:
    print(f"Error loading dataset: {e}")
    exit()

# Display the first 5 rows
print("First 5 Rows of the Dataset:")
print(iris.head())

# Check data types and missing values
print("\nDataset Info:")
print(iris.info())

print("\nMissing Values:")
print(iris.isnull().sum())

# Drop missing values if any
iris_cleaned = iris.dropna()

# Task 2: Basic Data Analysis
print("\nDescriptive Statistics:")
print(iris_cleaned.describe())

# Compute median for numerical columns
print("\nMedian Values:")
print(iris_cleaned.median(numeric_only=True))

# Group by species and compute the mean
grouped_means = iris_cleaned.groupby('species').mean(numeric_only=True)
print("\nMean of Numerical Columns Grouped by Species:")
print(grouped_means)

# Task 3: Data Visualization
sns.set(style="whitegrid")

# Line Chart (simulate trend over species)
plt.figure(figsize=(8, 4))
sns.lineplot(x='species', y='sepal_length', data=iris_cleaned, ci=None)
plt.title('Average Sepal Length by Species')
plt.xlabel('Species')
plt.ylabel('Sepal Length')
plt.tight_layout()
plt.savefig("line_chart.png")
plt.show()

# Bar Chart
plt.figure(figsize=(8, 4))
sns.barplot(x='species', y='petal_length', data=iris_cleaned)
plt.title('Average Petal Length by Species')
plt.xlabel('Species')
plt.ylabel('Petal Length')
plt.tight_layout()
plt.savefig("bar_chart.png")
plt.show()

# Histogram
plt.figure(figsize=(8, 4))
sns.histplot(iris_cleaned['sepal_width'], kde=True, bins=20)
plt.title('Distribution of Sepal Width')
plt.xlabel('Sepal Width')
plt.ylabel('Frequency')
plt.tight_layout()
plt.savefig("histogram.png")
plt.show()

# Scatter Plot
plt.figure(figsize=(8, 4))
sns.scatterplot(x='sepal_length', y='petal_length', hue='species', data=iris_cleaned)
plt.title('Sepal Length vs Petal Length by Species')
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.legend(title='Species')
plt.tight_layout()
plt.savefig("scatter_plot.png")
plt.show()

# Findings
print("\nFindings:")
print("- Iris-setosa has smaller petals compared to the other two species.")
print("- Sepal width appears normally distributed.")
print("- A positive correlation exists between sepal length and petal length.")

```
